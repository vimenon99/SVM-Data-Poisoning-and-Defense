```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import kneighbors_graph
from scipy.sparse.csgraph import connected_components
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc, precision_score, recall_score, f1_score

# Constants
RANDOM_SEED = 100
N_SAMPLES = 10000
PCA_COMPONENTS = 50
N_NEIGHBORS = 5

def load_and_preprocess_data():
    X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
    X, y = X[:N_SAMPLES], y[:N_SAMPLES].astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    normalizer = Normalizer()
    X_normalized = normalizer.transform(X_scaled)
    pca = PCA(n_components=PCA_COMPONENTS)
    X_pca = pca.fit_transform(X_normalized)

    # Construct feature similarity graph
    graph = kneighbors_graph(X_pca, N_NEIGHBORS, mode='connectivity', include_self=True)
    n_components, labels = connected_components(csgraph=graph, directed=False, return_labels=True)

    return X_pca, y, labels, n_components

# Detect and remove poisoned data using Isolation Forest
def remove_poisoned_data(X_train, y_train):
    isolation_forest = IsolationForest(random_state=RANDOM_SEED)
    anomaly_scores = isolation_forest.fit_predict(X_train)
    clean_mask = anomaly_scores != -1
    X_train_clean = X_train[clean_mask]
    y_train_clean = y_train[clean_mask]
    return X_train_clean, y_train_clean

# Function to perform label flipping attack
def label_flipping_attack(X_train, y_train, flip_ratio=0.1, random_state=42):
    np.random.seed(random_state)
    num_samples = len(y_train)
    num_flips = int(num_samples * flip_ratio)
    flip_indices = np.random.choice(num_samples, num_flips, replace=False)
    unique_classes = np.unique(y_train)
    flipped_labels = y_train.copy()
    for idx in flip_indices:
        current_label = flipped_labels[idx]
        possible_labels = unique_classes[unique_classes != current_label]
        new_label = np.random.choice(possible_labels)
        flipped_labels[idx] = new_label
    return X_train, flipped_labels

# Function to perform data injection attack
def data_injection_attack(X, y, injection_ratio=0.1, random_state=42):
    np.random.seed(random_state)
    num_samples = len(y)
    num_injections = int(num_samples * injection_ratio)
    unique_classes = np.unique(y)

    X_injected = []
    y_injected = []
    for _ in range(num_injections):
        random_class = np.random.choice(unique_classes)
        class_samples = X[y == random_class]
        random_sample = class_samples[np.random.choice(len(class_samples))]
        noise = np.random.normal(0, 0.1, size=X.shape[1])
        X_injected.append(random_sample + noise)
        y_injected.append(np.random.choice(unique_classes[unique_classes != random_class]))

    X_injected = np.array(X_injected)
    y_injected = np.array(y_injected)
    X_poisoned = np.concatenate([X, X_injected], axis=0)
    y_poisoned = np.concatenate([y, y_injected], axis=0)
    return X_poisoned, y_poisoned

def train_and_evaluate_svm(X, y, use_anomaly_detection=False, poison_attack=None, attack_params=None):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED)

    if poison_attack == 'label_flipping':
        X_train, y_train = label_flipping_attack(X_train, y_train, **attack_params)
    elif poison_attack == 'data_injection':
        X_train, y_train = data_injection_attack(X_train, y_train, **attack_params)

    if use_anomaly_detection:
        # Remove poisoned data using Isolation Forest
        X_train_clean, y_train_clean = remove_poisoned_data(X_train, y_train)
    else:
        X_train_clean, y_train_clean = X_train, y_train

    svm = SVC(kernel='rbf', C=10, gamma=0.01, probability=True, class_weight='balanced')
    svm.fit(X_train_clean, y_train_clean)

    y_pred = svm.predict(X_test)
    y_probs = svm.predict_proba(X_test)[:, 1]
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    if len(np.unique(y_test)) == 2:  # Binary classification
        tn, fp, fn, tp = conf_matrix.ravel()
        print(f"True Positives (TP): {tp}")
        print(f"False Positives (FP): {fp}")
        print(f"True Negatives (TN): {tn}")
        print(f"False Negatives (FN): {fn}")
    else:  # Multiclass classification
        print("Confusion Matrix:")
        print(conf_matrix)

    print(f"Accuracy: {accuracy:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")

    # Plot Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_probs, pos_label=1)
    pr_auc = auc(recall, precision)
    plt.figure()
    plt.plot(recall, precision, color='darkorange', lw=2, label=f'PR curve (area = {pr_auc:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend(loc="lower right")
    plt.show()


    # Plot confusion matrix heatmap
    plt.figure()
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.xticks(np.arange(len(np.unique(y))), np.unique(y))
    plt.yticks(np.arange(len(np.unique(y))), np.unique(y))
    plt.show()


    # Plot decision boundary (for binary classification and 2D feature spaces)
    if X_train_clean.shape[1] == 2:
        plt.figure()
        h = .02  # step size in the mesh
        x_min, x_max = X_train_clean[:, 0].min() - 1, X_train_clean[:, 0].max() + 1
        y_min, y_max = X_train_clean[:, 1].min() - 1, X_train_clean[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
        Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
        plt.scatter(X_train_clean[:, 0], X_train_clean[:, 1], c=y_train_clean, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.title('Decision Boundary')
        plt.show()
        return accuracy


    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_probs, pos_label=1)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

    return accuracy

# Main function
def main():
    X, y, _, _ = load_and_preprocess_data()

    # Standard SVM on clean data
    print("\nStandard SVM Results (Clean Data):")
    standard_svm_accuracy_clean = train_and_evaluate_svm(X, y, use_anomaly_detection=False)

    # Standard SVM on poisoned data (label flipping attack)
    print("\nStandard SVM Results (Poisoned Data - Label Flipping Attack):")
    flip_ratio = 0.1
    standard_svm_accuracy_poisoned_lf = train_and_evaluate_svm(X, y, use_anomaly_detection=False, poison_attack='label_flipping', attack_params={'flip_ratio': flip_ratio})

    # Standard SVM on poisoned data (data injection attack)
    print("\nStandard SVM Results (Poisoned Data - Data Injection Attack):")
    injection_ratio = 0.1
    standard_svm_accuracy_poisoned_di = train_and_evaluate_svm(X, y, use_anomaly_detection=False, poison_attack='data_injection', attack_params={'injection_ratio': injection_ratio})

    # SVM with anomaly detection on clean data
    print("\nSVM with Anomaly Detection Results (Clean Data):")
    svm_anomaly_detection_accuracy_clean = train_and_evaluate_svm(X, y, use_anomaly_detection=True)

    # SVM with anomaly detection on poisoned data (label flipping attack)
    print("\nSVM with Anomaly Detection Results (Poisoned Data - Label Flipping Attack):")
    svm_anomaly_detection_accuracy_poisoned_lf = train_and_evaluate_svm(X, y, use_anomaly_detection=True, poison_attack='label_flipping', attack_params={'flip_ratio': flip_ratio})

    # SVM with anomaly detection on poisoned data (data injection attack)
    print("\nSVM with Anomaly Detection Results (Poisoned Data - Data Injection Attack):")
    svm_anomaly_detection_accuracy_poisoned_di = train_and_evaluate_svm(X, y, use_anomaly_detection=True, poison_attack='data_injection', attack_params={'injection_ratio': injection_ratio})

    # Comparison
    print(f"\nComparison:")
    print(f"Standard SVM Accuracy (Clean Data): {standard_svm_accuracy_clean:.4f}")
    print(f"Standard SVM Accuracy (Poisoned Data - Label Flipping Attack): {standard_svm_accuracy_poisoned_lf:.4f}")
    print(f"Standard SVM Accuracy (Poisoned Data - Data Injection Attack): {standard_svm_accuracy_poisoned_di:.4f}")
    print(f"SVM with Anomaly Detection Accuracy (Clean Data): {svm_anomaly_detection_accuracy_clean:.4f}")
    print(f"SVM with Anomaly Detection Accuracy (Poisoned Data - Label Flipping Attack): {svm_anomaly_detection_accuracy_poisoned_lf:.4f}")
    print(f"SVM with Anomaly Detection Accuracy (Poisoned Data - Data Injection Attack): {svm_anomaly_detection_accuracy_poisoned_di:.4f}")
    
    
    
    # Plotting
    plt.figure(figsize=(15, 12))

    # Standard SVM Results
    plt.subplot(2, 3, 1)
    plt.bar("Clean Data", standard_svm_accuracy_clean, color='blue')
    plt.ylim(0.8, 1)
    plt.title("Standard SVM Accuracy (Clean Data)")

    plt.subplot(2, 3, 2)
    plt.bar("Label Flipping Attack", standard_svm_accuracy_poisoned_lf, color='red')
    plt.ylim(0.8, 1)
    plt.title("Standard SVM Accuracy (Label Flipping Attack)")

    plt.subplot(2, 3, 3)
    plt.bar("Data Injection Attack", standard_svm_accuracy_poisoned_di, color='red')
    plt.ylim(0.8, 1)
    plt.title("Standard SVM Accuracy (Data Injection Attack)")

    # SVM with Anomaly Detection Results
    plt.subplot(2, 3, 4)
    plt.bar("Clean Data", svm_anomaly_detection_accuracy_clean, color='blue')
    plt.ylim(0.8, 1)
    plt.title("SVM with Anomaly Detection Accuracy (Clean Data)")

    plt.subplot(2, 3, 5)
    plt.bar("Label Flipping Attack", svm_anomaly_detection_accuracy_poisoned_lf, color='red')
    plt.ylim(0.8, 1)
    plt.title("SVM with Anomaly Detection Accuracy (Label Flipping Attack)")

    plt.subplot(2, 3, 6)
    plt.bar("Data Injection Attack", svm_anomaly_detection_accuracy_poisoned_di, color='red')
    plt.ylim(0.8, 1)
    plt.title("SVM with Anomaly Detection Accuracy (Data Injection Attack)")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
